```
import os

train_file = open("train_augmented.txt", "w")
test_file = open("test_augmented.txt", "w")

current_path = os.getcwd()
print(current_path)

dir_list = os.listdir(current_path)
dir_list.sort()

for dir_name in dir_list:
    if os.path.isdir(dir_name):
        dir = current_path + "/" + dir_name
        print(dir)
        counter = 0
        path_list_unsorted=os.listdir(dir)
        path_list = sorted(path_list_unsorted) 

        for file_name in path_list:
            print(dir + '/' + file_name)
            
            if file_name[0] == '_':
              if file_name[-3:] == 'jpg' or file_name[-4:] == 'jpeg':
                  train_file.write(dir + '/' + file_name + '#')
                  train_file.close
                  continue
              else:
                  train_file.write(dir + '/' + file_name + "\n")
                  train_file.close
                  continue
            
            if counter < 320:
                if file_name[-3:] == 'jpg' or file_name[-4:] == 'jpeg':
                    train_file.write(dir + '/' + file_name + '#')
                    train_file.close
                else:
                    train_file.write(dir + '/' + file_name + "\n")
                    train_file.close
                    counter = counter + 1

            else:
                if file_name[-3:] == 'jpg' or file_name[-4:] == 'jpeg':
                    #print("writing11")
                    test_file.write(dir + '/' + file_name + '#')
                    #print("writing12")
                    test_file.close
                else:
                    test_file.write(dir + '/' + file_name + "\n")
                    #print("writing21")
                    test_file.close
                    counter = counter + 1
                    #print("writing22")
                
```

```
read_train = open("train_augmented.txt", "r").readlines()

with open("train_augmented.csv", mode="w", newline="") as train_file:
    for line in read_train:
        text_file = line.split("#")[-1].replace("\n", "")
        image_file = line.split('#')[0]
        #text_file = image_file.replace(".jpg", ".txt")
        data = [image_file, text_file]
        writer = csv.writer(train_file)
        writer.writerow(data)

read_test = open("test_augmented.txt", "r").readlines()

with open("test_augmented.csv", mode="w", newline="") as train_file:
    for line in read_test:
        #image_file = line.split("#")[-1].replace("\n", "")
        #text_file = image_file.replace(".jpg", ".txt")
        text_file = line.split("#")[-1].replace("\n", "")
        image_file = line.split('#')[0]
        data = [image_file, text_file]
        writer = csv.writer(train_file)
        writer.writerow(data)
```
